```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif
```

```python
calc_vif(dataset[["mpg","cylinders"]])
```

#### Independant Sample- Unpaired T Test

```python
dataset[dataset['name']=="ford torino"]["origin"]
```

```python
from scipy.stats import ttest_ind
dataset= dataset.dropna()
name = dataset[dataset['mpg'] == 'ford torino']['model_year']
name1 = dataset[dataset['mpg'] == 'vw pickup']['model_year']

ttest_ind(name, name1)
```

#### Dependant Sample-Paired T_Test
Same Group(Male) but Different Condition(ssc_p,hsc_p)


```python
from scipy.stats import ttest_rel
#dataset=dataset.dropna()
male = dataset[dataset['gender']=='M']['ssc_p']
male1 = dataset[dataset['gender']=='M']['hsc_p']
ttest_rel(male, male1)
```
